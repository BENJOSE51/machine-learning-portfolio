```python
# 0) Imports and settings
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, StackingRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.decomposition import PCA
from xgboost import XGBRegressor

# Reproducibility
RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)

# Plot settings
%matplotlib inline
sns.set_style('whitegrid')
```

```python
# 1) Load data
TRAIN_CSV = "../data/train.csv"
STORE_CSV = "../data/store.csv"

train = pd.read_csv(TRAIN_CSV, parse_dates=['Date'])
store = pd.read_csv(STORE_CSV)

print('train shape:', train.shape)
train.head()
```

```python
print('store shape:', store.shape)
store.head()
```

```python
# 2) Merge & initial checks

df = train.merge(store, on='Store', how='left')
print('Merged shape:', df.shape)

# Basic info and quick stats
print('\nDataFrame info:')
df.info()

print('\nSales & Customers description:')
print(df[['Sales','Customers']].describe())

# Null counts
null_counts = df.isnull().sum().sort_values(ascending=False)
print('\nColumns with nulls:')
print(null_counts[null_counts>0])
```

**Interpretation:**

- We merged `train` and `store` to get store-level metadata.
- The `Date` column is parsed as `datetime` for easy feature extraction.
- Reviewing `info()` and `describe()` helps identify datatype issues, obvious anomalies, and whether nulls exist.
- Next, we'll apply simple, defensible imputation and cleaner dtypes for categorical columns.

```python
# 3) Data cleaning & dtype fixes
cat_cols = ['StateHoliday','StoreType','Assortment','Promo_interval']
for c in cat_cols:
    if c in df.columns:
        df[c] = df[c].astype('category')

# Convert numeric-like columns safely
num_maybe = ['CompetitionOpenSinceMonth','CompetitionOpenSinceYear','Promo2SinceWeek','Promo2SinceYear']
for c in num_maybe:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors='coerce')

print('Dtypes after conversion:')
print(df.dtypes.value_counts())
```

**Interpretation:**

- Converting textual categories to `category` reduces memory usage and makes later encoding obvious.
- Numeric-like columns that contain NaNs or malformed values are coerced to numeric; we'll impute next.

```python
# 4) Imputation

# CompetitionDistance -> median
if 'CompetitionDistance' in df.columns:
    df['CompetitionDistance'] = df['CompetitionDistance'].fillna(df['CompetitionDistance'].median())

# Fill promo/competition
for c in ['CompetitionOpenSinceMonth','CompetitionOpenSinceYear','Promo2SinceWeek','Promo2SinceYear']:
    if c in df.columns:
        df[c] = df[c].fillna(0).astype(int)

# Promo_interval
if 'Promo_interval' in df.columns:
    df['Promo_interval'] = df['Promo_interval'].cat.add_categories(['None']).fillna('None')

print('Nulls after imputation:')
print(df.isnull().sum()[df.isnull().sum()>0])
```

**Interpretation:**

- Using median for `CompetitionDistance` is defensible when missingness likely means unknown distance.
- Promotion-related fields default to 0 or 'None' so that absence of information is treated as no promotion.
- These are simple, interview-friendly choices; in a production setting you might explore missingness patterns further.

```python
# 5) Outlier handling
num_cols = df.select_dtypes(include=['float64','int64']).columns.tolist()
exclude = ['Sales','Customers','Store','Date','CompetitionOpenSinceYear','CompetitionOpenSinceMonth','Promo2SinceWeek','Promo2SinceYear']
for x in exclude:
    if x in num_cols:
        num_cols.remove(x)

print('Numeric cols to cap (sample):', num_cols[:10])

def iqr_cap(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return series.clip(lower=lower, upper=upper)

for col in num_cols:
    df[col] = iqr_cap(df[col])

print('Applied IQR capping to numeric columns.')
```

**Interpretation:**

- IQR capping prevents extreme outliers from overly influencing model training while keeping most data intact.
- We chose capping rather than dropping rows to preserve data volume and maintain a simpler pipeline for an interview setting.

```python
# 6) Feature engineering:
# dateparts

df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df['DayOfWeek'] = df['Date'].dt.dayofweek

# Week of year
try:
    df['WeekOfYear'] = df['Date'].dt.isocalendar().week.astype(int)
except Exception:
    df['WeekOfYear'] = df['Date'].dt.week

# Open flag
df['Open'] = df['Open'].fillna(1).astype(int)

# Competition start and age months
if 'CompetitionOpenSinceYear' in df.columns and 'CompetitionOpenSinceMonth' in df.columns:
    df['CompetitionOpenSinceYear'] = df['CompetitionOpenSinceYear'].replace(0, np.nan)
    df['CompetitionOpenSinceMonth'] = df['CompetitionOpenSinceMonth'].replace(0, np.nan)
    df['CompetitionStart'] = pd.to_datetime(dict(year=df['CompetitionOpenSinceYear'].fillna(0).astype(int),
                                                 month=df['CompetitionOpenSinceMonth'].fillna(1).astype(int),
                                                 day=1), errors='coerce')
    df['CompetitionAgeMonths'] = ((df['Date'].dt.year - df['CompetitionStart'].dt.year) * 12 + 
                                 (df['Date'].dt.month - df['CompetitionStart'].dt.month)).fillna(0).clip(lower=0)
else:
    df['CompetitionAgeMonths'] = 0

# Promo flags
df['Promo2'] = df['Promo2'].fillna(0).astype(int)

if 'Promo_interval' in df.columns:
    def is_promo_month(row):
        if row['Promo_interval'] == 'None':
            return 0
        months = row['Promo_interval'].split(',')
        try:
            month_nums = [datetime.strptime(m.strip(), '%b').month for m in months]
        except Exception:
            month_nums = []
        return 1 if row['Month'] in month_nums else 0
    df['IsPromoMonth'] = df.apply(is_promo_month, axis=1)
else:
    df['IsPromoMonth'] = 0

# Log target
df['Sales_log'] = np.log1p(df['Sales'])

print('Feature engineering completed.')
```

**Interpretation:**

- Date parts capture seasonality and weekly patterns that affect staffing and inventory.
- `IsPromoMonth` and `Promo2` capture promotional activity; `CompetitionAgeMonths` approximates competitive pressure.
- Log-transforming `Sales` stabilizes variance and makes model errors more symmetric; we will inverse-transform predictions for reporting.

## Observations:

This makes the graph very right-skewed.  
That usually happens in retail data because of holidays, special events, or big promotions.  
Since this kind of skew can confuse models (they try too hard to fit those big spikes), I decided to log-transform the sales values to make them more balanced.

---

### 2 Sales Distribution (After Log Transformation)
After taking the log of sales, the graph looks much smoother and more balanced.  
The extreme spikes got compressed, and the overall shape is closer to normal.  
This makes it easier for models like linear regression to learn patterns more evenly across different stores and dates.  
When I evaluate the model later, I’ll convert predictions back to the original scale so that results are easy to understand.

---

### 3 Average Sales by Day of the Week
Here I saw a clear weekly pattern.  
Some days — usually Saturdays — have higher sales, while others like Sundays or Mondays are lower.  
This tells me there’s a routine pattern in customer behavior.  
It’s a good reason to keep `DayOfWeek` as a feature because it helps the model understand when sales usually peak or drop.

---

### 4 Rolling 7-Day Average of Total Sales
The rolling average helps smooth out daily ups and downs and shows the overall trend over time.  
There are repeating peaks, which probably match holidays or major promotions.  
This means sales are not random — there’s seasonality and trend involved.  
That’s why I added features like `Month`, `WeekOfYear`, and `IsPromoMonth` so the model can learn these time-based changes.

---

### 5 Correlation Heatmap
From the heatmap, I noticed:
- Sales and customers have a strong positive relationship — more customers, more sales (which makes sense).  
- Promotions (`Promo` and `IsPromoMonth`) are linked to higher sales.

These patterns confirm that promo and competition features are important to include in the model.

### 6 Overall Takeaways
- Sales are not random — they follow patterns that depend on **time**, **promotions**, and **competition**.  
- Applying a log transformation makes the data easier for models to handle.

In short, the EDA shows that a model that understands **time patterns, promotions, and competition** will perform better at predicting daily store sales.


```python
# 8) Prepare dataset for modeling
feature_cols = [
    'Open','DayOfWeek','Promo','StateHoliday','SchoolHoliday',
    'CompetitionDistance','CompetitionAgeMonths','Promo2','IsPromoMonth',
    'Year','Month','WeekOfYear'
]
if 'StoreType' in df.columns:
    feature_cols.append('StoreType')
if 'Assortment' in df.columns:
    feature_cols.append('Assortment')

model_df = df[feature_cols + ['Sales','Sales_log']].copy()

# One-hot encode simple categoricals
categorical_for_dummies = [c for c in model_df.columns if model_df[c].dtype.name == 'category' or model_df[c].dtype == object]
model_df = pd.get_dummies(model_df, columns=categorical_for_dummies, drop_first=True)

# Filter to open days with sales > 0
model_df = model_df[(model_df['Open']==1) & (model_df['Sales']>0)]
print('Model DF shape after filters and dummies:', model_df.shape)
```

```python
# 9) Train/test split
X = model_df.drop(columns=['Sales','Sales_log'])
y = model_df['Sales_log']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=RANDOM_STATE)

# Scaling for linear regression
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# PCA
MAX_COMPONENTS_FOR_PCA = 20
use_pca = False
if X_train.shape[1] > MAX_COMPONENTS_FOR_PCA:
    use_pca = True
    pca = PCA(n_components=min(MAX_COMPONENTS_FOR_PCA, X_train.shape[1]), random_state=RANDOM_STATE)
    X_train_pca = pca.fit_transform(X_train_scaled)
    X_test_pca = pca.transform(X_test_scaled)

print('X_train shape:', X_train.shape, 'X_test shape:', X_test.shape)
```

```python
# 10) Modeling:
results = []

def eval_and_store(name, y_true_log, y_pred_log):
    y_true = np.expm1(y_true_log)
    y_pred = np.expm1(y_pred_log)
    # Calculate RMSE manually without using squared parameter
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)  # Calculate square root manually
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    results.append({'model': name, 'RMSE': rmse, 'MAE': mae, 'R2': r2})
    print(f"{name} -> RMSE: {rmse:.2f}, MAE: {mae:.2f}, R2: {r2:.3f}")

# Linear Regression
lr = LinearRegression()
if use_pca:
    lr.fit(X_train_pca, y_train)
    y_pred_lr = lr.predict(X_test_pca)
else:
    lr.fit(X_train_scaled, y_train)
    y_pred_lr = lr.predict(X_test_scaled)

eval_and_store('LinearRegression', y_test, y_pred_lr)

# Random Forest
rf = RandomForestRegressor(n_estimators=100, random_state=RANDOM_STATE, n_jobs=-1)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
eval_and_store('RandomForest', y_test, y_pred_rf)

# Gradient Boosting
gb = GradientBoostingRegressor(n_estimators=100, random_state=RANDOM_STATE)
gb.fit(X_train, y_train)
y_pred_gb = gb.predict(X_test)
eval_and_store('GradientBoosting', y_test, y_pred_gb)

# XGBoost
xgb = XGBRegressor(n_estimators=200, random_state=RANDOM_STATE, verbosity=0)
xgb.fit(X_train, y_train)
y_pred_xgb = xgb.predict(X_test)
eval_and_store('XGBoost', y_test, y_pred_xgb)

# Stacking
estimators = [
    ('rf', RandomForestRegressor(n_estimators=50, random_state=RANDOM_STATE)),
    ('gb', GradientBoostingRegressor(n_estimators=50, random_state=RANDOM_STATE))
]
final_estimator = GradientBoostingRegressor(n_estimators=100, random_state=RANDOM_STATE)
stack = StackingRegressor(estimators=estimators, final_estimator=final_estimator, n_jobs=-1, passthrough=False)
stack.fit(X_train, y_train)
y_pred_stack = stack.predict(X_test)
eval_and_store('StackingRegressor', y_test, y_pred_stack)
```

**Metrics Used:**
- **RMSE (Root Mean Squared Error):** Measures how far predictions are from actual sales — lower is better.  
- **MAE (Mean Absolute Error):** Average absolute difference between predicted and actual sales.  
- **R² (R-squared):** Shows how much of the sales variation the model can explain — higher is better.


### Model Comparison

After training all models, I compared their performance using **RMSE**, **MAE**, and **R²** scores.  
The table above shows that the more advanced ensemble models performed noticeably better than the simpler ones.

**Observations:**
- **Linear Regression** serves as a good baseline, but its performance is weaker because it can’t fully capture the non-linear relationships between features like promotions, competition, and seasonal effects.
- **Random Forest** performs much better since it handles non-linear patterns and feature interactions automatically.
- **Gradient Boosting** and **XGBoost** improve further by learning from previous errors — boosting models usually excel on structured, tabular data like this.
- **Stacking Regressor** gave the **best results overall**, combining the strengths of multiple models and slightly improving accuracy compared to individual learners.

**Interpretation:**
- The results make sense — retail sales data is complex, with interactions between date, competition, and promotions.  
  Ensemble methods like **XGBoost** and **Stacking** capture these patterns far better than a single linear model.
- A higher **R²** value and lower **RMSE/MAE** mean the ensemble models are more accurate and generalize well.
