# Loan Approval Prediction

**Goal:** Predict whether a loan application is approved (`Loan_Status = Y/N`) using applicant information.

**Notebook flow:**
1. Load data  
2. Quick checks (`info`, missing values, `describe`)  
3. EDA (target distribution, numeric histograms, correlation heatmap, unique values)  
4. Simple feature engineering  
5. Simple imputation & encoding (`get_dummies`)  
6. Train/test split and modeling (Logistic Regression + Random Forest)  
7. Evaluation and feature importance  
8. Short business takeaways

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    roc_auc_score, confusion_matrix, classification_report
)

RANDOM_STATE = 42
dp = "../data/train.csv"
sns.set(style='whitegrid')
```

## Load the dataset

```python
df = pd.read_csv(dp)
print('Data shape:', df.shape)
df.head()
```

## Quick info & missing values

```python
df.info()
df.isnull().sum().sort_values(ascending=False)
```

## Numeric summary

```python
df.describe().T
```

## Target mapping & cleaning

```python
df['Loan_Status'] = df['Loan_Status'].map({'Y': 1, 'N': 0})
if 'Dependents' in df.columns:
    df['Dependents'] = df['Dependents'].replace('3+', 3)
    df['Dependents'] = pd.to_numeric(df['Dependents'], errors='coerce')
```

## Numeric vs categorical columns

```python
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
num_cols = [c for c in num_cols if c != 'Loan_Status']
cat_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()
print('Numeric columns:', num_cols)
print('Categorical columns:', cat_cols)
```

## Check unique values for categorical columns

```python
# Display unique values for all categorical columns
for col in cat_cols:
    print(f'\nUnique values in {col}:')
    print(df[col].unique())
```

```python
plt.figure(figsize=(5,3))
sns.countplot(x='Loan_Status', data=df)
plt.title('Target distribution (0 = Not approved, 1 = Approved)')
plt.show()
```

### Interpretation
The target variable is slightly imbalanced — most applicants have their loans approved (`1`).  
This means the model should focus on recall, ensuring we correctly predict approvals without missing too many.


## Numeric feature distributions

```python
for col in ['LoanAmount', 'ApplicantIncome', 'CoapplicantIncome']:
    if col in df.columns:
        plt.figure(figsize=(6,3))
        sns.histplot(df[col].dropna(), bins=30, kde=True)
        plt.title(f'{col} distribution')
        plt.show()
```

### Interpretation
`LoanAmount` and `ApplicantIncome` are both right-skewed.  
Most applicants earn and borrow smaller amounts, while a few have much higher values.  
This skewness is normal for income-related data.


## Correlation heatmap

```python
if len(num_cols) >= 2:
    plt.figure(figsize=(8,6))
    corr = df[num_cols].corr()
    sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', vmin=-1, vmax=1)
    plt.title('Correlation heatmap (numeric features)')
    plt.show()
```

### Interpretation
There’s only a mild positive correlation between `ApplicantIncome` and `LoanAmount`.  
No pair of features shows very strong correlation, meaning most variables contribute unique information to the model.


## Simple feature engineering

```python
if set(['ApplicantIncome','CoapplicantIncome']).issubset(df.columns):
    df['Total_Income'] = df['ApplicantIncome'].fillna(0) + df['CoapplicantIncome'].fillna(0)
if 'LoanAmount' in df.columns and 'Loan_Amount_Term' in df.columns:
    df['EMI'] = df['LoanAmount'] / (df['Loan_Amount_Term'].replace(0, np.nan))
```

## Simple missing-value handling

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
numeric_cols = [c for c in numeric_cols if c != 'Loan_Status']
for c in numeric_cols:
    df[c] = df[c].fillna(df[c].median())
for c in cat_cols:
    if c in df.columns:
        mode_val = df[c].mode().iloc[0] if not df[c].mode().empty else 'Missing'
        df[c] = df[c].fillna(mode_val)
```

## Prepare features and encode categoricals

```python
if 'Loan_ID' in df.columns:
    df = df.drop(columns=['Loan_ID'])
y = df['Loan_Status']
X = df.drop(columns=['Loan_Status'])
X_encoded = pd.get_dummies(X, drop_first=True)
print('Feature matrix shape after encoding:', X_encoded.shape)
```

## Train-test split

```python
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, stratify=y, random_state=RANDOM_STATE)
print('Train shape:', X_train.shape)
print('Test shape:', X_test.shape)
print('Train target distribution:\n', y_train.value_counts(normalize=True))
```

## Model training

```python
log = LogisticRegression(random_state=RANDOM_STATE, max_iter=1000)
log.fit(X_train, y_train)
rf = RandomForestClassifier(n_estimators=200, random_state=RANDOM_STATE)
rf.fit(X_train, y_train)
```

## Evaluation

```python
def evaluate(model, X_test, y_test, name='Model'):
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:,1] if hasattr(model, 'predict_proba') else None
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, zero_division=0)
    rec = recall_score(y_test, y_pred, zero_division=0)
    f1 = f1_score(y_test, y_pred, zero_division=0)
    roc = roc_auc_score(y_test, y_proba) if y_proba is not None else np.nan
    print(f'\\n=== {name} ===')
    print(f'Accuracy: {acc:.4f}')
    print(f'Precision: {prec:.4f}')
    print(f'Recall: {rec:.4f}')
    print(f'F1-score: {f1:.4f}')
    if not np.isnan(roc):
        print(f'ROC-AUC: {roc:.4f}')
    print('\\nClassification report:\\n', classification_report(y_test, y_pred, digits=4))
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(4,3))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f'{name} Confusion Matrix')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

evaluate(log, X_test, y_test, 'Logistic Regression')
evaluate(rf, X_test, y_test, 'Random Forest')
```

## Feature importance

```python
fi = pd.Series(rf.feature_importances_, index=X_encoded.columns).sort_values(ascending=False)
fi.head(15)
plt.figure(figsize=(6,4))
fi.head(15).plot(kind='barh')
plt.gca().invert_yaxis()
plt.title('Top 15 Features (Random Forest)')
plt.show()
```

## Conclusion & next steps

- Logistic Regression provides a good baseline; Random Forest improves performance.
- Credit history, total income, and loan amount are usually strong predictors.
- Next steps: hyperparameter tuning, cross-validation, SHAP for interpretability.
